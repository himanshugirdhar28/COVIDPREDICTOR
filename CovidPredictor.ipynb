```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
import math
```

```python
def getPixels(filename):
    img = Image.open(filename, 'r')
    w, h = img.size
    pix = list(img.getdata())
    return [pix[n:n+w] for n in range(0, w*h, w)]
```

```python
def data_fetch(d,m):
    l=[]
    j=1
    for i in range(1,m+1):
        rgb_matrix=getPixels(d+str(j)+'.jpg')
        l.append(rgb_matrix)
        j+=1
    return(l)
```

```python
def brightness_matrix(sample_rgb_list):
    #brightness matrix for each image
    sample_rgb_list_brightness={}
    l=1
    for i in sample_rgb_list:
        sample_rgb_list_brightness[l]=[]
        for j in range(len(i)):
            sample_rgb_list_brightness[l].append([])
            for k in i[j]:
                # print(k[0],k[1],k[2],(int(k[0])+int(k[1])+int(k[2])),(k[0]+k[1]+k[2])/3)
                sample_rgb_list_brightness[l][j].append((int(k[0])+int(k[1])+int(k[2]))/3)  #   (0.299*float(k[0])+0.587*float(k[1])+0.114*float(k[2]))
        # break
        l+=1
    return(sample_rgb_list_brightness)
```

```python
def brightness_matrix_value(sample_rgb_list_brightness,m):
    sample_rgb_list_brightness_value={}
    for i in np.arange(1,m+1):
        sum=0
        count=0
        for j in sample_rgb_list_brightness[i]:
            for k in j:
                sum+=k
                count+=1
        sample_rgb_list_brightness_value[i]=(sum/count)
    return(sample_rgb_list_brightness_value)
```

```python
def rgb_matrix_value(sample_rgb_list,m):
    sample_rgb_list_value={}
    for i in np.arange(m):
        sum1=0
        sum2=0
        sum3=0
        count=0
        for j in sample_rgb_list[i]:
            for k in j:
                sum1+=k[0]**2
                sum2+=k[1]**2
                sum3+=k[2]**2
                count+=1
        sample_rgb_list_value[i+1]=[(sum1/count)**0.5,(sum2/count)**0.5,(sum3/count)**0.5]
    return(sample_rgb_list_value)
```

Model Building to export

```python
class knn:
    def __init__(self):
        self.pos = None
        self.neg = None

    def fit(self, pos_list, neg_list):
        self.pos=pos_list
        self.neg=neg_list

    def predict(self, i):
        distance_vector_pos=[]
        distance_vector_neg=[]
        for j in self.pos:
            distance_vector_pos.append(math.dist(i,j))
        for j in self.neg:
            distance_vector_neg.append(math.dist(i,j))
        distance_vector_pos.sort()
        distance_vector_neg.sort()
        count_pos=0
        count_neg=0
        x=0
        y=0
        k=5
        while(count_pos+count_neg<k):
            if(distance_vector_pos[x]<distance_vector_neg[y]):
                x+=1
                count_pos+=1
            else:
                y+=1
                count_neg+=1
        if(count_pos>count_neg):
            result_knn="positive"
        else:
            result_knn="negative"
        return(result_knn)
```

```python
model=knn()
model.fit(pos_neg_rgb_matrix_values_tr['p'],pos_neg_rgb_matrix_values_tr['n'])
```
